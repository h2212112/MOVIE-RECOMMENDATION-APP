```python

def convert_with_string(obj):
    obj = ast.literal_eval(obj)
    L = []
    for i in obj:
        L.append(i['name'])
    return L

def convert_without_string(obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L

```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview']= movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['tags'] = movies['overview']+ movies['genres']+ movies['keywords']+movies['cast']+ movies['crew']
```

```python
new_df = movies[['movie_id','title', 'tags']]
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
# Applying stemming
from nltk.stem.porter import PorterStemmer 

ps = PorterStemmer()

def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
new_df
```

### Vectorizing 

```python
from sklearn.feature_extraction.text import CountVectorizer

# Extracting most frequently use 5000 features.  
cv  = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
new_df
```

```python

# Calculating the angular distance between vectors (cos θ)

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
# Making recommendation function
def recomended(movie):
    recomended_movie_list = []
    recomended_movie_list_index = []
    # getting movie index
    movie_index = new_df[new_df['title']== movie].index[0]
    # Getting distances from all the movies 
    distances = similarity[movie_index]
    # Getting top 5 movies
    movies_list = sorted(list(enumerate(distances)),reverse=True, key= lambda x: x[1])[1:6]

    for i in movies_list:
         recomended_movie_list_index.append(i[0])
         
#     Getting Movie name
    for i in movies_list:
         recomended_movie_list.append((new_df.iloc[i[0]].title))
         
    return  recomended_movie_list_index, recomended_movie_list
        
    
```

```python
recomended('Bad Grandpa')
```

```python
def recommended2(movie_id):
    recommended_movie_list_index = []
    
   
    if movie_id not in movies['movie_id'].values:
        raise ValueError(f"Movie ID {movie_id} not found in the dataset.")
    
    movie_index = movies[movies['movie_id'] == movie_id].index[0]
    
    distances = similarity[movie_index]
    
   
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        recommended_movie_list_index.append(new_df.iloc[i[0]].movie_id)
    
    return recommended_movie_list_index
```
